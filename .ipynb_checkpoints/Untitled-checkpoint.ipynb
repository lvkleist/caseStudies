```python
import numpy as np
sigma = np.array((1, 0.6, 0.3,0.6, 1, 0.4, 0.3, 0.4, 1))
sigma = sigma.reshape((3, 3))
totalLoss = np.zeros(100000)

data = pd.read_csv('joined_transformed_v2.csv')
data = data.sample(n=1000)
```

```python

for i in range(100000):
    if(i%1000 == 0): print(i)
    totalLoss[i] = computeTotalLoss(data)
    
```

```python
totalLoss[90:100]
```

```python
def computeTotalLoss(data):
    z = np.random.multivariate_normal([0, 0, 0], sigma)
    df = data.copy()
    df['eps'] = np.random.normal(size = len(df['Loan']))
    df['Xi'] = df['Alpha']*z[df['Region']] + df['Gamma']*df['eps']
    
    df['Default'] = df['Xi'] < df['Threshold']
    df['Loss'] = df['LGD']*df['EAD']*df['Default']
    return np.sum(df['Loss'])
    
    
    
```

```python
bins = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550]
cats = pd.cut(totalLoss, bins)
```
